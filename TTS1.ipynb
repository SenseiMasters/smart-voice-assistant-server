```python
from TTS.config import load_config
from TTS.utils.manage import ModelManager
from TTS.utils.synthesizer import Synthesizer
```

```python
config="persian-tts-female-vits/config.json"
model="persian-tts-female-vits/best_model_30824.pth"

model_path =model # Absolute path to the model checkpoint.pth
config_path =config # Absolute path to the model config.json
```

```python
# text="سلام، به همه دانشجویان عزیز ورودتان را به کلاس مباحث ویژه نرم افزار تبریک می گویم"
text="ز گهواره تا گور طلب دانوش"

synthesizer = Synthesizer(
    model_path, config_path
)
wavs = synthesizer.tts(text)
synthesizer.save_wav(wavs, 'sp.wav')
```
